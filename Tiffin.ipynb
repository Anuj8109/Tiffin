```python
import re
from datetime import datetime, time
from openpyxl import Workbook

def filter_whatsapp_chat(filename, sender_name, month, keyword):
    wb = Workbook()

    # Select the active worksheet
    ws = wb.active
    # Open the WhatsApp chat export file

    ws.title = "My Data"

    # Add some data
    ws.append(['Date', 'Lunch', 'Dinner'])
    for i in range(1,32):
      ws.append([str(i),'N','N'])
    with open(filename, 'r', encoding='utf-8') as file:
        lines = file.readlines()

    # Define a list to hold filtered messages
    filtered_messages = []

    # Loop through each line in the file
    afternoon_start = time(12, 0)  # 12 PM
    afternoon_end = time(18, 0)  # 6PM
    for line in lines:
        match = re.match(r'^\[(\d{1,2}/\d{1,2}/\d{2,4}), (\d{1,2}:\d{2}:\d{2}\s[APMapm]{2})\] (.*?): (.*)', line)

        if match:
            # Extract components from the matched pattern
            date_str, time_str, sender, message = match.groups()

            # Parse the date using dd/mm/yyyy format
            try:
                date_obj = datetime.strptime(date_str, "%d/%m/%y")
            except ValueError:
                continue  # Skip lines with unrecognized date formats

            # Parse the time
            time_obj = datetime.strptime(time_str, "%I:%M:%S %p")  # 12-hour format with AM/PM
            # Check if the message meets all filter conditions
            if (sender == sender_name and
                date_obj.strftime("%B") == month and
                re.search(r'\b' + re.escape(keyword) + r'\b', message, re.IGNORECASE)):
                # Add to filtered messages list

                day = int(date_obj.strftime('%d'))
                try:
                    day = int(date_obj.strftime('%d'))
                    patterns = [
                        r'(\d+)\s*tiffin',          # Matches "2 tiffin" or "2tiffin"
                        r'tiffin\s*(\d+)',           # Matches "tiffin 2"
                        r'(\d+)\s*[Tt]iffins',       # Matches "2 tiffins"
                        r'[Tt]iffin\s*[Nn]umber\s*:\s*(\d+)' # Matches "Tiffin Number: 2"
                    ]
                    count = 0
                    for pattern in patterns:
                      match = re.search(pattern, message, re.IGNORECASE)
                      if match:
                          count = int(match.group(1))
                    if afternoon_start <= time_obj.time() < afternoon_end:
                        # Mark dinner column for afternoon times
                        ws.cell(row=day + 1, column=3, value=str(count))
                        filtered_messages.append(f"{date_str} {time_str} - {sender}: {message} -> Night")
                        #print("Night")
                    else:
                        # Mark lunch column for other times
                        filtered_messages.append(f"{date_str} {time_str} - {sender}: {message} -> Lunch")
                        ws.cell(row=day+2, column=2, value=str(count))

                except Exception as e:
                    print(f"Error processing date {date_str}: {e}")

    wb.save('Tiffin_Count.xlsx')



    return filtered_messages

# Usage example
filename = 'chat.txt'
filtered_messages = filter_whatsapp_chat(filename, sender_name="Naman Gandhinagar", month="November", keyword="Tiffin")

# Display the filtered messages
print("Total Tiffin " + str(len(filtered_messages)))
for message in filtered_messages:
    print(message)
```
